```python
link_start = 'https://realpython.github.io/fake-jobs/jobs/'
link_end = '.html'
```

```python
i = 0

for each_title in titles_list:
    each_title = each_title.replace(' ', '-')
    each_title = each_title.replace('(', '')
    each_title = each_title.replace(')', '')
    each_title = each_title.replace("'", '')
    each_title = each_title.replace('/', '-')
    each_title = each_title.replace(',', '')
    print(link_start + each_title.lower() + '-' + str(i) + link_end)
    i+=1
```

```python
clean_link_list = []

for i, each_title in enumerate(titles_list):
    each_title = each_title.replace(' ', '-')
    each_title = each_title.replace('(', '')
    each_title = each_title.replace(')', '')
    each_title = each_title.replace("'", '')
    each_title = each_title.replace('/', '-')
    each_title = each_title.replace(',', '')
    clean_link_list.append(link_start + each_title.lower() + '-' + str(i) + link_end)

clean_link_list
```

```python
jobs['Apply'] = clean_link_list
```
